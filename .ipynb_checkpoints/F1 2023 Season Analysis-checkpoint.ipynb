```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
from datetime import timedelta
%matplotlib inline
```

```python
calendar = pd.read_csv('formula1_2023season_calendar.csv')
calendar.set_index('Round', inplace=True)
drivers = pd.read_csv('formula1_2023season_drivers.csv')
drivers.set_index('Abbreviation', inplace=True)
teams = pd.read_csv('formula1_2023season_teams.csv')
teams.index = range(1,11)
sprintResults = pd.read_csv('formula1_2023season_qualifyingResults.csv')
raceResults = pd.read_csv('formula1_2023season_raceResults.csv')
```

```python
def assign_color(val_type, values):
    color = []
    for val in values:
        if val_type == 'drivers':  abbr = val.split()[1].upper()[0:3]
        elif val_type == 'teams':  abbr = val[0:4].upper()
        if abbr in ['ALFA','RAI','GIO','KUB']:   cl.append('#780a0a')
        elif abbr in ['HAAS','SCH','MAZ']:       cl.append('#ffffff')
        elif abbr in ['ASTO','VET','STR']:       cl.append('#006f62')
        elif abbr in ['WILL','RUS','LAT']:       cl.append('#0072ff')
        elif abbr in ['ALPH','GAS','TSU']:       cl.append('#2b5962')
        elif abbr in ['MCLA','RIC','NOR']:       cl.append('#ff9721')
        elif abbr in ['RED ','VER','PER']:       cl.append('#070480')
        elif abbr in ['FERR','LEC','SAI']:       cl.append('#fc0808')
        elif abbr in ['MERC','HAM','BOT']:       cl.append('#07f5dd')
        elif abbr in ['ALPI','ALO','OCO']:       cl.append('#30a5ff')
    return color
```

```python
driver_standings_df = raceResults.groupby('Driver')['Points'].sum().reset_index()
driver_standings_df = driver_standings_df.sort_values(by='Points', ascending=False)

# Reset index to make it more readable and start rankings from 1
driver_standings_df.reset_index(drop=True, inplace=True)
driver_standings_df.index += 1
# Display the driver standings
display(driver_standings_df)
```

```python
# Styling preferences
plt.style.use('ggplot')
plt.style.use('dark_background')
plt.rcParams['axes.facecolor'] = '#15151d'
plt.rcParams['figure.facecolor'] = '#15151d'
plt.rcParams['grid.color'] = '#444444'

# Extract unique tracks for x-axis labels
tracks = raceResults['Track'].unique()

# Top 10 drivers based on standings
top10 = driver_standings_df['Driver'].values[:10]

# Function to assign colors - here we use a simple method for demonstration
def assign_color(category, items):
    colors = plt.cm.tab10(np.linspace(0, 1, len(items)))
    return colors

# Assign colors to top 10 drivers
c = assign_color("drivers", top10)

# Plot settings
plt.figure(figsize=(16,7))
plt.axis([-0.4, len(tracks) - 0.6, -10, max(driver_standings_df['Points']) + 50])

# Loop through top 10 drivers to plot their points progression
for i, driver in enumerate(top10):
    ls = '-'
    # Extract driver's race-by-race points and calculate cumulative sum
    pts = raceResults[raceResults['Driver'] == driver]['Points'].values
    # Example adjustments (modify or remove based on actual data/requirements)
    if driver == "Max Verstappen":     pts = np.append(pts, [3, 2, 2]) # Example adjustment
    elif driver == "Lewis Hamilton":   pts = np.append(pts, [2]) # Example adjustment
    # Add more conditions here if needed for the 2023 season
    
    # Plot cumulative points
    plt.plot(np.cumsum(pts), linewidth=3, label=driver.split()[1], color=c[i], linestyle=ls)

# Add title, labels, legends with custom styling
plt.title("Formula 1 - 2023 Season\nTop 10 Drivers' Points Progression", color='#bbbbbb', fontsize=22, fontweight='bold')
plt.legend(fontsize=15)
plt.xlabel('RACES', fontsize=16, color='#ffff55', fontweight='bold')
plt.xticks(range(0, len(tracks)), tracks, fontsize=14, color='#ffff55', rotation=80)
plt.ylabel('POINTS', fontsize=16, color='#ffff55', fontweight='bold')
plt.yticks(fontsize=14, color='#ffff55')
```

```python

# Get the cumulative points for each driver after each race
cumulative_points_df = raceResults.pivot_table(values='Points', index='Track', columns='Driver', aggfunc='sum').fillna(0).cumsum()

# Filter out the top 2 drivers (Verstappen and Perez) and calculate the points gap
top_drivers = cumulative_points_df[['Max Verstappen', 'Sergio Perez']]
points_gap = top_drivers['Max Verstappen'] - top_drivers['Sergio Perez']

# Set up the plot
plt.figure(figsize=(14, 8))

# Create bars for the points gap, color blue for Verstappen and orange for Perez
bar_colors = ['blue' if gap > 0 else 'orange' for gap in points_gap]
bars = plt.bar(points_gap.index, points_gap, color=bar_colors)

# Add value labels on bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval if yval > 0 else yval - 5, int(yval), 
             ha='center', va='bottom' if yval > 0 else 'top', color='white')

# Add titles and labels
plt.title('Points Gap Between Max Verstappen and Sergio Perez - Formula One 2023 Season', fontsize=14, color='white')
plt.xlabel('Races', fontsize=12, color='white')
plt.ylabel('Points Gap', fontsize=12, color='white')
plt.xticks(rotation=90, color='white')
plt.yticks(color='white')

# Set face colors
plt.gca().set_facecolor('#15151d')
plt.gcf().set_facecolor('#15151d')

# Show plot
plt.tight_layout()
plt.show()
```

In the graph, the bars are colored based on the **points gap** between **Max Verstappen and Sergio Perez**. A bar is colored blue if the gap is positive, which means that Verstappen is ahead in points. If Perez had been leading at any point, those bars would be colored orange.

Since all the bars are blue, it indicates that Max Verstappen has been leading Sergio Perez in terms of cumulative points after each race throughout the season, and there were no races where Perez had a cumulative lead over Verstappen. Therefore, no orange bars are present in the graph.

If there would have been any races where Perez was ahead in cumulative points, those races would have bars pointing downwards (below the zero line on the y-axis) and colored orange to indicate Perez's lead.

```python
# Calculate constructor standings
# Sum the points of both drivers for each team across all races
constructor_standings_df = raceResults.groupby('Team')['Points'].sum().sort_values(ascending=False).reset_index()

# Plotting the constructor standings as a horizontal bar chart
plt.figure(figsize=(14, 8))
bars = plt.barh(constructor_standings_df['Team'], constructor_standings_df['Points'], color='skyblue')
plt.xlabel('Total Points')
plt.title('Constructor Standings - Formula One 2023 Season')
plt.gca().invert_yaxis()  # Highest points at the top

# Add the data labels on the bars
for bar in bars:
    plt.text(bar.get_width() - (bar.get_width()*0.05), bar.get_y() + bar.get_height()/2, 
             f'{bar.get_width():.0f}', 
             va='center', ha='right', color='black', fontsize=12)

# Show plot
plt.show()
```
